```python
# importa a classe pyplot da biblioteca matplotlib como plt
import matplotlib.pyplot as plt

import numpy as np
```

```python
# configura o matplotlib para plotar os resultados in line
%matplotlib inline
```

```python
# criando amostra de dados
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
# forma Funcional do matplotlib (menos controle dos elementos gráficos)
plt.plot(x, y)

# descrição do eixo x e y e Título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
# alterando a cor e o formato da linha
plt.plot(x, y, 'r')
```

```python
plt.plot(x, y, 'r--')
```

```python
# plot com dois ou mais gráficos
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
# utilizando objetos figure  e axes
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')
axes2.plot(y, x)
```

```python
# criando objetos com subplots

fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
# o objeto axe é uma lista de axes
ax
```

```python
# inserindo elemento nos axes por iteração
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('Título')
```

```python
fig
```

```python
# pode-se também inserir elementos nos axes por vez
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, y, 'b')
ax[0].set_title('título 1')

ax[1].plot(y, x, 'g')
ax[1].set_title('título 2')
```

```python
## acertando layout e evitando sobreposições com tight_layout()

fig, ax = plt.subplots(nrows=5, ncols=5)
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig = plt.figure(figsize = (8, 4), dpi=100)
```

```python
## alterando o tamanho da figura

fig, axes = plt.subplots(figsize = (12, 3))

axes.plot(x, y, 'r-')
axes.set_title('Título')
```

```python
# salvando a figura
fig.savefig('imagem.jpg')
fig.savefig('imagem.png')
fig.savefig('imagem.pdf')
```

```python
## adicionando mais de um plot no mesmo axe

fig, axes = plt.subplots(figsize = (5,8))

axes.plot(x, y, 'r-', label = 'Dados 1')
axes.plot(y, x, 'g--', label = 'Dados 2')
axes.set_title('Título')

## adicionando legenda
axes.legend()
```
