### Plots categóricos

Agora vamos discutir como usar seaborn para traçar dados categóricos. Existem alguns tipos de argumentos principais para isso:

- factorplot
- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

Apresentaremos exemplos de cada um:

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# barplot - mostra o valor médio de uma variável por categoria
# o traço preto indica o desvio padrão da variável

sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# pode-se trocar a função, ao invés de valor médio exibir outras coisas
# no caso abaixo, usaremos o desvio padrão

sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# countplot - é um barplot para contagem das amostras por categoria
sns.countplot(x='sex', data=tips)
```

```python
# boxplot
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
# pode-se ainda quebrar cada categoria (dia) em subcategorias (sexo)
# com o parâmetro hue
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex')
```

```python
# orientação do boxplot

sns.boxplot(data=tips, palette='rainbow', orient='h')
```

```python
# violinplot
sns.violinplot(x='day', y='tip', data=tips)
```

```python
# hue também funciona
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
```

```python
# pra desploluir o gráfico pode-se utilizar o parâmetro split
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
```

```python
# stripplot
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
# para melhorar a noção de densidade utiliza-se o parâmetro jitter
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
# pode-se utilizar o hue
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
# pode-se utilizar o split pra separar as categorias
sns.stripplot(x='day', y='total_bill', data=tips, 
              jitter=True, hue='sex', split=True)
```

```python
# swarmplot - dá ideia de densidade
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# utilizando categorização dos dados com o parâmetro hue
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# despoluindo com split
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# sugestão, swarm plot com violin plot

sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# factorplot - permite que se crie qualquer tipo de plot via parâmetro

sns.factorplot('sex', y='total_bill', data=tips, kind='swarm')
```

```python
sns.factorplot('sex', y='total_bill', data=tips, kind='violin')
```

```python
sns.factorplot('sex', y='total_bill', data=tips, kind='bar')
```
