### Cores e estilos

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# darkgrid
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips)
```

```python
# whitegrid
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
# ticks
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
```

```python
# white
sns.set_style('white')
sns.countplot(x='sex', data=tips)
```

```python
# dark
sns.set_style('dark')
sns.countplot(x='sex', data=tips)
```

```python
# tirar linhas
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
# tirar linhas
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True, bottom=True)
```

```python
# alterando o tamanho da figura
sns.set_style('ticks')
plt.figure(figsize=(10,8))
sns.countplot(x='sex', data=tips)
```

```python
# alterando tamanho com parametros do seaborn

sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# alterando tamanho com parametros do seaborn

sns.lmplot(x='total_bill', y='tip', data=tips, size=4, aspect=2)
```

```python
# set_context
sns.set_context('poster')
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# set_context
sns.set_context('paper')
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# set_context
sns.set_context('talk')
sns.lmplot(x='total_bill', y='tip', data=tips)
```

### Paleta de cores

```python
# set_paleta
sns.set_context('notebook')
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',
           palette='inferno')
```
