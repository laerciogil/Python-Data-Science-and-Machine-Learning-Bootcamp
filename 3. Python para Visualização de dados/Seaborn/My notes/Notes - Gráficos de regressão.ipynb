### Gráficos de regressão

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# lmplot - Linear Model
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# categorizando as regressões
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
# alterando a paleta de cores
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',
          palette='rainbow')
```

```python
# alterando o tipo de marcadores
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',
          markers=['o','v'])
```

```python
# parâmetro scatter
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',
          markers=['o','v'], scatter_kws={'s':100})
```

```python
# separando em axes por linhas e colunas

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex')
```

```python
# separando em mais de uma variável categórica
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex',
          row='time')
```

```python
# aspect e size
# separando em mais de uma variável categórica
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day',
          aspect=0.6, size=8)
```
